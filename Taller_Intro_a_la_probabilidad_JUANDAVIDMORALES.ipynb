## Primer punto Distribuciones discretas.
1. Cree un arreglo de N muestras con N a su selección, pero mayor a 1000, de números distribuidos con las correspondientes distribuciones, para cada arreglo de números: 
1. Grafique los números aleatorios, 
2. Ordene de menor a mayor y grafique,
4. Encuentre los valores teoricos de media y varianza, y comparelos con los estimados.

```
#Para todas las simulaciones:
N=1000000 #Se aumentaron los puntos N para poder visualizar mejor los histogramas
m=200
numpy.random.seed(seed=2**32 - 1)#Para fijar una semilla para los experimentos y que los numeros generados siempre sean los mismos en cada corrida completa de este fuente.
```

```
# bernoulli con cierto p entre 0 y 1
p=0.3
from scipy.stats import bernoulli
mean, var, skew, kurt = bernoulli.stats(p, moments='mvsk')
r = bernoulli.rvs(p, size=N)
#ver https://docs.scipy.org/doc/scipy-0.14.0/reference/stats.html
plt.plot(r)
plt.title(str(N)+" puntos distribuidos bernoulli con p="+str(p))
plt.show()


```

```
r.sort()
plt.plot(r)
plt.title(str(N)+" puntos distribuidos bernoulli ordenados con p="+str(p))
plt.show()
```

```
plt.hist(r, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos bernoulli con p="+str(p))
plt.show()
print("Note que hasta ahora no sabemos el tipo de nuestro arreglo r:")
print(type(r))
print("Vamos a hallar algunas estimaciones de estadisticas de r")
# ver https://numpy.org/doc/stable/reference/routines.statistics.html 
avg=r.mean()
varest=r.var()
print("---ESTIMACIONES----")
print("Los valores teoricos de la media y varianza para la bernoulli con p=",p,"son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza para la bernoulli con p=",p,"son u_est=" ,avg,"var_est=",varest)
```

```
# binomial con cierto p entre 0 y 1 y cierto n
n,p = 10, 0.5
from scipy.stats import binom
mean1, var1, skew1, kurt1 = binom.stats(n, p, moments='mvsk')
r1 = binom.rvs(n, p, size=N)
plt.plot(r1)
plt.title(str(N)+" puntos distribuidos de manera binomial con p="+str(p)+" y n= "+str(n))
plt.show()

r1.sort()
plt.plot(r1)
plt.title(str(N)+" puntos distribuidos de dist binomial ordenados con p="+str(p)+" y n= "+str(n))
plt.show()

plt.hist(r1, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos binomialmente")
plt.show()

print("---ESTIMACIONES----")
avg1=r1.mean()
varest1=r1.var()
print("Los valores teoricos de la media y varianza para la dist binomial con p=",p," y n= ",n,"son u=" ,mean1,"var=",var1)
print("Los valores estimados en nuestro experimento de la media y varianza para la dist binomial con p=",p," y n= ",n,"son u_est=" ,avg1,"var_est=",varest1)

```

```
# Geométrica con cierto p entre 0 y 1
p = 0.3
from scipy.stats import geom
mean2, var2, skew2, kurt2 = geom.stats(p, moments='mvsk')
r2 = geom.rvs(p, size=N)
plt.plot(r2)
plt.title(str(N)+" puntos distribuidos de manera geometrica con p="+str(p))
plt.show()

r2.sort()
plt.plot(r2)
plt.title(str(N)+" puntos distribuidos de dist geometrica ordenados con p="+str(p))
plt.show()

plt.hist(r2, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos geometricamente")
plt.show()

print("---ESTIMACIONES----")
avg2=r2.mean()
varest2=r2.var()
print("Los valores teoricos de la media y varianza para la dist geometrica con p=",p,"son u=" ,mean2,"var=",var2)
print("Los valores estimados en nuestro experimento de la media y varianza para la dist geometrica con p=",p,"son u_est=" ,avg2,"var_est=",varest2)
```

```
# Poisson con cierto Lamnda L
L = 5
from scipy.stats import poisson
mean3, var3, skew3, kurt3 = poisson.stats(L, moments='mvsk')
r3 = poisson.rvs(L, size=N)
plt.plot(r3)
plt.title(str(N)+" puntos distribuidos por Poisson con Lamnda="+str(L))
plt.show()

r3.sort()
plt.plot(r3)
plt.title(str(N)+" puntos distribuidos de dist poisson ordenados con Lamnda="+str(L))
plt.show()

plt.hist(r3, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos por Poisson")
plt.show()

print("---ESTIMACIONES----")
avg3=r3.mean()
varest3=r3.var()
print("Los valores teoricos de la media y varianza para la dist geometrica con Lamnda=",L,"son u=" ,mean3,"var=",var3)
print("Los valores estimados en nuestro experimento de la media y varianza para la dist geometrica con Lamnda=",L,"son u_est=" ,avg3,"var_est=",varest3)
```

```
#Binomial negativa con parametro p después de r fallos
r,p = 5,0.5
from scipy.stats import nbinom
mean4, var4, skew4, kurt4 = nbinom.stats(r, p, moments='mvsk')
r4 = nbinom.rvs(r, p, size=N)
plt.plot(r4)
plt.title(str(N)+" puntos distribuidos por binomial negativa con r="+str(r)+" y p= "+str(p))
plt.show()

r4.sort()
plt.plot(r4)
plt.title(str(N)+" puntos distribuidos de dist binomial negativa ordenados con r="+str(r)+" y p= "+str(p))
plt.show()

plt.hist(r4, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos por dist binomial negativa")
plt.show()

print("---ESTIMACIONES----")
avg4=r4.mean()
varest4=r4.var()
print("Los valores teoricos de la media y varianza para la dist binom negativa con r=",r," y p= ",p," son u=" ,mean4,"var=",var4)
print("Los valores estimados en nuestro experimento de la media y varianza para la dist binom negativa con r=",r," y p= ",p,"son u_est=" ,avg4,"var_est=",varest4)
```

## Segundo punto Distribuciones contínuas.


```
#Para todas las simulaciones:
N=200000
m=300
numpy.random.seed(seed=1314926535)#Para fijar una semilla para los experimentos y que los numeros generados siempre sean los mismos en cada corrida completa de este fuente.
```

```
# Normal de media 5 y desviación estandar 0.5
u=5
std=0.5
mean, var, skew, kurt = scipy.stats.norm.stats(moments='mvsk',loc=u,scale=std)
r = scipy.stats.norm.rvs (loc=u,scale=std, size=N)
#ver https://docs.scipy.org/doc/scipy-0.14.0/reference/stats.html
plt.plot(r)
plt.title(str(N)+" puntos distribuidos normales con media="+str(u)+" y std="+str(std))
plt.show()
r.sort()
plt.plot(r)
plt.title(str(N)+" puntos distribuidos normales ordenados con p="+str(p))
plt.show()
plt.hist(r, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos normales con u="+str(u)+ " y std= "+str(std))
plt.show()
print("arreglo r:")
print(type(r))

print("---ESTIMACIONES----")
avg=r.mean()
varest=r.var()
print("Los valores teoricos de la media y varianza para la normal son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza son u_est=" ,avg,"var_est=",varest)
```

```
# Uniforme entre 0 y 255
a=0
b=255
mean5, var5, skew5, kurt5 = scipy.stats.uniform.stats(loc=a,scale=b,moments='mvsk')
r5 = scipy.stats.uniform.rvs (loc=a,scale=b,size=N)
plt.plot(r5)
plt.title(str(N)+" puntos distribuidos uniformemente entre 0 y 255")
plt.show()
r5.sort()
plt.plot(r5)
plt.title(str(N)+" puntos distribuidos uniformemente ordenados")
plt.show()
plt.hist(r5, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos uniformemente")
plt.show()
print("arreglo r:")
print(type(r5))

print("---ESTIMACIONES----")
avg5=r5.mean()
varest5=r5.var()
print("Los valores teoricos de la media y varianza para la dist uniforme son u=" ,mean5,"var=",var5)
print("Los valores estimados en nuestro experimento de la media y varianza son u_est=" ,avg5,"var_est=",varest5)
```

```
# Exponencial con Lamnda L
L=0.3
mean6, var6, skew6, kurt6 = scipy.stats.expon.stats(scale=1/L, moments='mvsk')
r6 = scipy.stats.expon.rvs (scale=1/L,size=N)
plt.plot(r6)
plt.title(str(N)+" puntos distribuidos exponencialmente con Lamnda="+str(L))
plt.show()
r6.sort()
plt.plot(r6)
plt.title(str(N)+" puntos distribuidos exponencialmente y ordenados con Lamnda="+str(L))
plt.show()
plt.hist(r6, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos exponencialmente")
plt.show()
print("arreglo r:")
print(type(r6))

print("---ESTIMACIONES----")
avg6=r6.mean()
varest6=r6.var()
print("Los valores teoricos de la media y varianza para la dist exponencial son u=" ,mean6,"var=",var6)
print("Los valores estimados en nuestro experimento de la media y varianza son u_est=" ,avg6,"var_est=",varest6)
```

```
#Chi cuadrada con K grados de libertad
K=3.0

mean7, var7, skew7, kurt7 = scipy.stats.chi2.stats(K, moments='mvsk')
r7 = scipy.stats.chi2.rvs (K,size=N)
plt.plot(r7)
plt.title(str(N)+" puntos distribuidos por Chi 2 con grados de libertad="+str(K))
plt.show()
r7.sort()
plt.plot(r7)
plt.title(str(N)+" puntos distribuidos por Chi 2 y ordenados con grados de libertad="+str(K))
plt.show()
plt.hist(r7, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos por Chi 2")
plt.show()
print("arreglo r:")
print(type(r7))

print("---ESTIMACIONES----")

avg7=r7.mean()
varest7=r7.var()
print("Los valores teoricos de la media y varianza para la dist Chi 2 son u=" ,mean7,"var=",var7)
print("Los valores estimados en nuestro experimento de la media y varianza son u_est=" ,avg7,"var_est=",varest7)
```

```
#Gamma con parametros k y theta t
k,t=2.54,0.43

mean8, var8, skew8, kurt8 = scipy.stats.gamma.stats(k,scale = (1/(1/t)), moments='mvsk')
r8 = scipy.stats.gamma.rvs (k,scale = (1/(1/t)),size=N)
plt.plot(r8)
plt.title(str(N)+" puntos distribuidos por dist gamma con k="+str(k)+"y theta= "+str(t))
plt.show()
r8.sort()
plt.plot(r8)
plt.title(str(N)+" puntos distribuidos por dist gamma ordenados con k="+str(k)+"y theta= "+str(t))
plt.show()
plt.hist(r8, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos por dist gamma")
plt.show()
print("arreglo r:")
print(type(r8))

print("---ESTIMACIONES----")

avg8=r8.mean()
varest8=r8.var()
print("Los valores teoricos de la media y varianza para la dist gamma son u=" ,mean8,"var=",var8)
print("Los valores estimados en nuestro experimento de la media y varianza son u_est=" ,avg8,"var_est=",varest8)
```
